
### Question 1 — Despine & Font Size

The plot above has **small fonts**, unnecessary **spines**, and no axis labels.

1. Improve the figure by:
   * Removing the top and right spines.

```python
np.random.seed(0)
categories = ['A', 'B', 'C', 'D']
values = np.random.randint(5, 15, size=len(categories))

#Fixed Code
fig, ax = plt.subplots(figsize=(4, 3))
ax.bar(categories, values, color='lightgrey')
ax.set_title('Sales', fontsize=10)   

#Remove the top and right spines
sns.despine()

#Increase title font size
ax.set_title("Sales", fontsize = 18)

#Increase tick labels font size
ax.tick_params(axis = "x", labelsize = 13)
ax.tick_params(axis = "y", labelsize = 13)

#Add axis labels
ax.set_xlabel("Category", fontsize = 13)
ax.set_ylabel("Number of sales", fontsize = 13)

plt.show()
```

### Report 1:


### Question 3 — Over-plotting & Transparency

The scatter plot over-plots points, making dense regions hard to perceive.

1. Mitigate over-plotting using **transparency (`alpha`)**, smaller marker size.
2. [Challenge only] Use a color palette consistent with data density through ax.hexbin(x, y, cmap = 'ARGUMENT'), where you provide argument
3. Provide a short reflection on how these adjustments clarify patterns.


```python
np.random.seed(3)
n = 500
x = np.random.normal(size=n)
y = 2 * x + np.random.normal(size=n)

fig, ax = plt.subplots(figsize=(4, 3))
#Make it transparent 80% transparent
ax.scatter(x, y, color = 'red', s = 20, alpha = 0.2) 
ax.set_title('Scatter', fontsize = 9)
plt.show()

fig2, ax2 = plt.subplots(figsize = (4,3))
#Challenge - data density
hexbin = ax2.hexbin(x, y, gridsize = 50, cmap = 'viridis') 
fig2.colorbar(hexbin, ax = ax2, label = 'Counts')
ax2.set_title("Scatter (with data density)", fontsize = 12)
plt.show()
```

### Report 3:

2. The scatter plot with data density provides a better representation of how the data is spread across the dataset. Overall, these changes allow the viewer to see patterns from the scatter plot (more effectively).


### Question 4 — Axes & Legend Placement

The time-series plot suffers from **cramped x-axis labels** and a legend that overlaps tick labels.

1. Rotate date ticks, use automatic date formatting, and expand the figure width.
2. Move the legend to a clearer position or outside the plotting area.
3. Explain why these revisions improve interpretability.


```python
import matplotlib.dates as mdates

np.random.seed(4)
dates = pd.date_range('2025-01-01', periods=100, freq='D')
series1 = np.cumsum(np.random.randn(100))
series2 = np.cumsum(np.random.randn(100))

#Expand width
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(dates, series1, label='Series 1')
ax.plot(dates, series2, label='Series 2')
ax.set_title('Time Series', fontsize=9)

#Rotate data ticks and use automatic date formatting
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.setp(ax.get_xticklabels(), rotation = 0, ha = "center", fontsize = 10)

#Move the legend to a clearer position
ax.legend(loc = 'lower right')   #or loc = 'best'
plt.show()
```

### Report 4:
1. Expanding the width of the graph allows the viewer to see the lines better as they are not clumped together. Rotating the date ticks and increasing the font sizes improve readability. 
2. Moving the legend to the corner (where it is clear) improves readability too.
3. These changes, overall, improve interpretability because they allow the viewer to clearly see the pattern over time. 

```python
import matplotlib.dates as mdates

fig, ax = plt.subplots(figsize=(8, 4), dpi=120)
ax.plot(dates, series1, label='Series 1', linewidth=1.5)
ax.plot(dates, series2, label='Series 2', linewidth=1.5)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Cumulative Sum', fontsize=12)
ax.set_title('Random Walk Time-Series', fontsize=14, weight='bold')
ax.legend(loc='upper left', frameon=False)
ax.xaxis.set_major_locator(mdates.MonthLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))
plt.setp(ax.get_xticklabels(), rotation=45, ha='right')
fig.tight_layout()
plt.show()
```


### Question 5 — Histogram Bin Choice & Labels

The histogram uses **100 bins**, creating a noisy appearance.

1. Select an appropriate bin width using a rule (e.g., Freedman-Diaconis) or domain knowledge (your domain knowledge in this case can be trial and error).
2. Choose colors with sufficient contrast and remove heavy edges.
3. Add axis labels and a descriptive title.
4. Reflect on how your binning choice balances detail and clarity.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

np.random.seed(5)
data = np.random.gamma(shape=2., scale=1.5, size=1000)

fig, ax = plt.subplots(figsize=(4, 3))

# Freedman–Diaconis rule: h = 2 * IQR / n^(1/3)
IQR = np.percentile(data, 75) - np.percentile(data, 25)
bin_width = 2 * IQR / (len(data) ** (1/3))

num_bins = math.ceil((data.max() - data.min()) / bin_width) #Get number of bins from bin_width

ax.hist(data, bins = num_bins, color = 'blue') #Removed edgecolor

#Change title and axis label
ax.set_title('Distribution of Apple Prices', fontsize = 9)
ax.set_xlabel("Apple Prices")
ax.set_ylabel("Frequency")

#Remove spine
sns.despine()

plt.tight_layout()
plt.show()
```

### Report 5:
1. For the bin_width, I decided to use the Freedman-Diaconis rule to compute the best bin_width for this dataset. I also despined and removed heavy edges like the edgecolor. Finally, I added axis labels and an informative title. 
2. These changes balance detail and clarity as the labels and title provide more information about the graph and its distribution while the appropriate bin_width makes the graph look more clear for the viewers. Removing the heavy edges also did this as there were too many lines in the first graph which made the graph look ugly.
